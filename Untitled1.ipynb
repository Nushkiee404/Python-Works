```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Specified')
df['country'] = df['country'].fillna('Unknown')

df = df.dropna(subset=['date_added'])
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors = 'coerce')
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df[['date_added','year_added']].head()
```

```python
yearly_counts = df['year_added'].value_counts().sort_index()
yearly_counts
```

```python
country_df = df['country'].str.split(', ').explode()
```

```python
top_countries = country_df.value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values,y=top_countries.index)
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
genre_df = df['listed_in'].str.split(', ').explode()
```

```python
top_genres = genre_df.value_counts().head(10)
top_genres
```

```python
movies_df = df[df['type'] == 'Movie']
```

```python
movies_df['duration_minutes'] = movies_df['duration'].str.contains('min')
```

```python
movies_df = movies_df.dropna(subset=['duration'])
```

```python
movies_df = movies_df[movies_df['duration'].str.contains('min')]
```

```python
movies_df = movies_df[movies_df['duration'].str.contains('min', na=False)]
```

```python
movies_df['duration_minutes'] = (
    movies_df['duration']
    .str.replace(' min', '', regex=False)
    .astype(int)
)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(movies_df['duration_minutes'], bins=30, kde=True)
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

```python
titles_per_year = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(12,5))
sns.lineplot(x=titles_per_year.index, y=titles_per_year.values)
plt.title("Netflix Content Growth Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)
plt.title("Distribution of Content Ratings on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Countries Producing Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.subplot(2,2,2)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="magma")
plt.title("Top 10 Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.subplot(2,2,3)
sns.histplot(movies_df['duration_minutes'], bins=30, kde=True, color="skyblue")
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")

plt.subplot(2,2,4)
sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index,
    palette="coolwarm"
)
plt.title("Content Rating Distribution")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")

plt.tight_layout()
plt.show()
```
